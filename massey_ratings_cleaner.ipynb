All data pulled from: https://www.masseyratings.com/data

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import re
```

```python
master_df = pd.DataFrame(columns=['Team1', 'Team1_Score', 'Team2', 'Team2_Score', 'Winner', 'Year'])

# Iterate through files and cleanup
for i in range(2013, 2019):
    
    # Read in file
    year = str(i)
    test = pd.read_fwf('unclean_tournament_results/' + year + '_results.csv', header=None)

    # Drop unneeded columns
    test = test[[1, 2, 3, 4]]

    # Rename columns
    test.rename(columns={1: 'Team1', 2: 'Team1_Score', 3: 'Team2', 4: 'Team2_Score'}, inplace=True)

    # Drop @ symbol
    test['Team1'] = test['Team1'].str.replace(r'@', '')
    test['Team2'] = test['Team2'].str.replace(r'@', '')

    # Strip whitespace
    test['Team1'] = test['Team1'].str.strip()
    test['Team2'] = test['Team2'].str.strip()

    # Determine winner
    test['Winner'] = ['Team1' if int(x) > int(y) else 'Team2'\
                        for x, y in zip(test['Team1_Score'], test['Team2_Score'])]

    # Create Year Column
    test['Year'] = year
    
    master_df = pd.concat([master_df, test], ignore_index=True)
    
master_df.to_csv('tournament_results/2013_2018.csv', index=False)
master_df
```
